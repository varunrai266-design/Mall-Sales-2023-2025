```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Repair\Downloads\mall_sales 2023-2025.csv")
df.head()
```

```python
 df.describe()
```

```python
df.isnull().sum()
```

```python
mean_value = df['Unit Price'].mean()
df['Unit Price'] = df['Unit Price'].fillna(mean_value).round().astype('int64')
```

```python
df.head()
```

```python
mean_value = df['Units Sold'].mean()
df['Units Sold'] = df['Units Sold'].fillna(mean_value).round().astype('int64')
```

```python
df.head(15)
```

```python
df['Customer Name'] = df['Customer Name'].ffill()
df['Sales Channel'] = df['Sales Channel'].ffill()
```

```python
df.head(15)
```

```python
import duckdb

# Write your query pointing directly to the 'df' variable
sql_query = """
SELECT 
    "Product",
    "Unit Price",
    "Units Sold",
    ("Unit Price" * "Units Sold") AS "Total Sale"
FROM 
    df;
"""

# Run the query using DuckDB's direct connection execution
result_df = duckdb.query(sql_query).df()
result_df.head()
```

```python
df['Total Sale'] = (df['Unit Price'] * df['Units Sold']).round().astype('int64')

# Check the results to make sure the NaNs are gone
df.head(15)
```

```python
df['Customer Name'] = df['Customer Name'].bfill()
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Product'] = df['Product'].str.lower()

# Check the changes
df.head()
```

```python
state_counts = df['State'].value_counts()

print(state_counts)
```

```python
state_sales = df.groupby('State')['Total Sale'].sum().reset_index()
state_sales = state_sales.sort_values(by='Total Sale', ascending=False)
state_sales
```

```python
df['Sale Date'] = pd.to_datetime(df['Sale Date'], format='%d-%m-%Y')
df['Year'] = df['Sale Date'].dt.year

yearly_sales = df.groupby('Year')['Total Sale'].sum().reset_index()
yearly_sales = yearly_sales.sort_values(by='Year')
yearly_sales
```

```python
customer_sales = df[df['Customer Name'].notnull()].groupby('Customer Name')['Total Sale'].sum().reset_index()
customer_sales = customer_sales.sort_values(by='Total Sale', ascending=False)
top_10_customers = customer_sales.head(10)
top_10_customers
```

```python
# 1. Filter out missing names, group by Customer Name, and sum the Total Sale numbers
customer_revenue = df[df['Customer Name'].notnull()].groupby('Customer Name')['Total Sale'].sum().reset_index()

# 2. Sort the customers from the highest spender to the lowest
customer_revenue = customer_revenue.sort_values(by='Total Sale', ascending=False)

# Display the full summary table
customer_revenue
```

```python

product_sales = df.groupby('Product')['Total Sale'].sum().reset_index()

product_sales = product_sales.sort_values(by='Total Sale', ascending=False)

top_10_products = product_sales.head(10)

top_10_products
```

```python
# 1. Group by State and add up the Units Sold
state_units = df.groupby('State')['Units Sold'].sum().reset_index()

# 2. Sort from the highest units sold to the lowest
state_units = state_units.sort_values(by='Units Sold', ascending=False)

# Display the summary table
state_units
```

```python
# 1. Group by State and Product, then calculate the average Unit Price
state_product_pricing = df.groupby(['State', 'Product'])['Unit Price'].mean().reset_index()

# 2. Rename the column to reflect that it is an average
state_product_pricing.columns = ['State', 'Product', 'Avg Unit Price']

# 3. Sort alphabetically by State and then by highest price down
state_product_pricing = state_product_pricing.sort_values(by=['State', 'Avg Unit Price'], ascending=[True, False])

# Display the summary table
state_product_pricing
```

```python
# 1. Group by Sales Channel and calculate the sum for both columns
channel_summary = df.groupby('Sales Channel').agg(
    Total_Sales=('Total Sale', 'sum'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()

# 2. Sort from highest revenue-generating channel to lowest
channel_summary = channel_summary.sort_values(by='Total_Sales', ascending=False)

# Display the combined table
channel_summary
```

```python
# 1. Extract the Year and Month periods (creates a clean format like '2024-01', '2024-02')
df['Month_Year'] = df['Sale Date'].dt.to_period('M')

# 2. Group by the new Month_Year period and sum the Total Sale numbers
monthly_sales = df.groupby('Month_Year')['Total Sale'].sum().reset_index()

# 3. Sort chronologically from oldest month to newest month
monthly_sales = monthly_sales.sort_values(by='Month_Year')

# 4. Convert Month_Year back to a string format so it plays nice with charts and Excel
monthly_sales['Month_Year'] = monthly_sales['Month_Year'].astype(str)

# Display the summary table
monthly_sales
```

```python
# 1. Group by Product and count how many rows (transactions) exist for each name
product_counts = df.groupby('Product').size().reset_index(name='Total Transactions')

# 2. Sort from the most frequently bought product down to the least
product_counts = product_counts.sort_values(by='Total Transactions', ascending=False)

# Display the summary leaderboard
product_counts
```

```python
import os
import pandas as pd


# 1. Setup your secure target folder path
save_path = r"C:\Users\Repair\Desktop\mall sales"
os.makedirs(save_path, exist_ok=True)
excel_file = os.path.join(save_path, "final_sales_report.xlsx")

# 2. Make sure Sale Date is datetime format before running aggregations
df['Sale Date'] = pd.to_datetime(df['Sale Date'], format='%d-%m-%Y', errors='coerce')

# 3. Compile all your analytical reporting dataframes
monthly_sales         = df.groupby(df['Sale Date'].dt.to_period('M'))['Total Sale'].sum().reset_index()
channel_summary       = df.groupby('Sales Channel').agg(Total_Sales=('Total Sale', 'sum'), Total_Units_Sold=('Units Sold', 'sum')).reset_index()
state_product_pricing = df.groupby(['State', 'Product'])['Unit Price'].mean().reset_index()
state_units           = df.groupby('State')['Units Sold'].sum().reset_index()
top_10_products       = df.groupby('Product')['Total Sale'].sum().reset_index().sort_values(by='Total Sale', ascending=False).head(10)
customer_revenue      = df.groupby('Customer Name')['Total Sale'].sum().reset_index().sort_values(by='Total Sale', ascending=False)
top_10_customers      = customer_revenue.head(10)

# Format time period to string for clean Excel integration
monthly_sales['Sale Date'] = monthly_sales['Sale Date'].astype(str)

# 4. Save the entire workspace to Excel across organized spreadsheet tabs
with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
    # SAVES YOUR MAIN CLEAN DATA TABLE FROM THE SCREENSHOT
    df.to_excel(writer, sheet_name="master_clean_data", index=False)
    
    # SAVES ALL YOUR METRIC INSIGHTS
    monthly_sales.to_excel(writer, sheet_name="monthly_sales", index=False)
    channel_summary.to_excel(writer, sheet_name="channel_summary", index=False)
    state_product_pricing.to_excel(writer, sheet_name="state_product_pricing", index=False)
    state_units.to_excel(writer, sheet_name="state_units", index=False)
    top_10_products.to_excel(writer, sheet_name="top_10_products", index=False)
    customer_revenue.to_excel(writer, sheet_name="customer_revenue", index=False)
    top_10_customers.to_excel(writer, sheet_name="top_10_customers", index=False)

print("🎉 Complete success! Your master clean dataset and all 7 metrics are safely saved to your desktop!")
```
